##SSH Setting
백업한 id_rsa 파일을 로컬에 저장해야 함

```python
cd /content/drive/MyDrive/Colab/
```

```python
!mkdir -p ~/.ssh/
!cp id_rsa ~/.ssh/
!chmod 400 ~/.ssh/id_rsa
!ssh-keyscan -t rsa github.com >> ~/.ssh/known_hosts
!ssh -T git@github.com
```
